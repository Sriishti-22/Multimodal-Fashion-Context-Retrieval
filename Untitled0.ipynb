```python
COLORS = ["red","blue","white","black","yellow","green","grey","brown"]

CLOTHING = [
    "shirt","t-shirt","hoodie","blazer","jacket",
    "coat","dress","tie","pants","jeans"
]

ENVIRONMENTS = ["office","park","street","home","mall","outdoor"]

STYLES = ["formal","casual","business","sporty","weekend"]
```

```python
def encode_text(prompts):
    tokens = clip.tokenize(prompts).to(device)
    with torch.no_grad():
        return model.encode_text(tokens).cpu().numpy()

TEXT_EMBEDS = {
    "color": encode_text([f"a photo of {c} clothing" for c in COLORS]),
    "clothing": encode_text([f"a person wearing a {c}" for c in CLOTHING]),
    "environment": encode_text([f"a photo taken in a {e}" for e in ENVIRONMENTS]),
    "style": encode_text([f"a {s} outfit" for s in STYLES]),
}
```

```python
os.makedirs("embeddings", exist_ok=True)

image_vectors = []
image_ids = []

image_dir = "/content/data/images"

for img_name in tqdm(os.listdir(image_dir)):
    img_path = os.path.join(image_dir, img_name)

    image = preprocess(Image.open(img_path).convert("RGB")) \
            .unsqueeze(0).to(device)

    with torch.no_grad():
        img_feat = model.encode_image(image).cpu().numpy()[0]

    record = {"global": img_feat}

    for key in TEXT_EMBEDS:
        sims = img_feat @ TEXT_EMBEDS[key].T
        record[key] = TEXT_EMBEDS[key][sims.argmax()]

    image_vectors.append(record)
    image_ids.append(img_name)

np.save("embeddings/vectors.npy", image_vectors)
np.save("embeddings/image_ids.npy", image_ids)
```

```python
os.makedirs("indexes", exist_ok=True)

vectors = np.load("embeddings/vectors.npy", allow_pickle=True)

keys = ["global","color","clothing","environment","style"]

for key in keys:
    mat = np.stack([v[key] for v in vectors]).astype("float32")
    index = faiss.IndexFlatIP(mat.shape[1])
    index.add(mat)
    faiss.write_index(index, f"indexes/{key}.index")

print("FAISS indexes built successfully")
```

```python
image_ids = np.load("embeddings/image_ids.npy")

indexes = {
    k: faiss.read_index(f"indexes/{k}.index")
    for k in ["global","color","clothing","environment","style"]
}

WEIGHTS = {
    "global": 0.4,
    "color": 0.2,
    "clothing": 0.2,
    "environment": 0.1,
    "style": 0.1
}

def parse_query(query):
    parsed = {
        "global": [query],
        "color": [],
        "clothing": [],
        "environment": [],
        "style": []
    }

    query_lower = query.lower()

    for c in COLORS:
        if c in query_lower:
            parsed["color"].append(c)

    for cl in CLOTHING:
        if cl in query_lower:
            parsed["clothing"].append(cl)

    for e in ENVIRONMENTS:
        if e in query_lower:
            parsed["environment"].append(e)

    for s in STYLES:
        if s in query_lower:
            parsed["style"].append(s)

    return parsed

def search(query, top_k=5):
    parsed = parse_query(query)
    scores = np.zeros(len(image_ids))

    for key, terms in parsed.items():
        if not terms:
            continue

        tokens = clip.tokenize(terms).to(device)
        with torch.no_grad():
            emb = model.encode_text(tokens).mean(dim=0).cpu().numpy()

        emb = emb.astype("float32").reshape(1, -1)
        sims, idxs = indexes[key].search(emb, len(image_ids))
        scores[idxs[0]] += WEIGHTS[key] * sims[0]

    top = scores.argsort()[::-1][:top_k]
    return image_ids[top]
```

```python
def show_results(image_names, cols=5):
    rows = (len(image_names) + cols - 1) // cols
    plt.figure(figsize=(15, 3 * rows))

    for i, img_name in enumerate(image_names):
        img_path = os.path.join("/content/data/images", img_name)

        if not os.path.exists(img_path):
            continue

        img = Image.open(img_path).convert("RGB")
        plt.subplot(rows, cols, i + 1)
        plt.imshow(img)
        plt.title(img_name[:12])
        plt.axis("off")

    plt.show()
```

```python
queries = [
    "A person in a bright red raincoat",
    "Professional business attire inside a modern office",
    "Someone wearing a blue shirt sitting on a park bench",
    "Casual weekend outfit for a city walk",
    "A red tie and a white shirt in a formal setting"
]

for q in queries:
    print("\nQUERY:", q)
    results = search(q, top_k=5)
    show_results(results)
```
